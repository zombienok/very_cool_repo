```python
import numpy as np

# создание из обычных списков 
a = np.array(list(range(1, 6)))
a
```

```python
# создание массивов с использованием методов
# создаёт массив с shape (4, 4) из рандомных чисел
empty = np.empty([3, 3])
# создаёт матрицу вандермонда на основе последовательности чисел и степени
b = np.vander(a, 5)
empty, b[:5]
```

```python
# атрибут shape показывает размерность тензора
empty.shape, b.shape
```

```python
# dtype - тип массива
empty.dtype, b.dtype
```

```python
# различные варианты обращения по срезам в массиве
print(b[1, 1]) # 1-я строка, 1-й столбец (счёт с нуля)
print(b[1:]) #все строки, начиная с 1-й
print(b[[2,3], [2, 3]]) # 2-й элемент из 2-й и 3-й элемент из 3-й строки
```

```python
print(b[2:4, 2:4]) #2 по 3 элемент из 2 по 3 строк
print(b[:,2]) #2-й элемент из всех строк
```

```python
a = np.array([np.logspace(start=1, stop=2, num=5, base=np.exp(1)), np.logspace(start=1, stop=4, num=5, base=2)])
b = np.ones(a.shape) * 5
a, b
```

```python
# применение векторных операций 
print(a + b) # сложение
print(a - b) # вычитание
print(a * b) # поэлементное умножение
```

```python
#и операций из мат.статистики
print(a.mean()) # среднее значение по массиву
print(np.median(a)) # медиана
print(np.average(a, axis=0)) # среднее значение по столбцам
```
